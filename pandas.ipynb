```python
#MERGE FUNCTION SHOULD DEFINITELY HAVE SAME COLUMN NAME WITH SAME ITEMS IN EACH CORRESPONDING COLUMN
a=pd.DataFrame({"ab":[1,2,3,4,5],"ac":[12,445,345,56,8],"ad":[12,4,454,7,6]})
b=pd.DataFrame({"ab":[1,2,3,4,5],"ac":[12,445,345,56,8],"ad":[12,4,454,7,6]})
c=pd.merge(a,b)
print(c)
```

```python
#JOIN FUNCTION SHOULD DEFINITELY HAVE DIFFERENT COLUMN NAME
a=pd.DataFrame({"ab":[1,2,3,4,5],"ac":[12,445,345,56,8],"ad":[12,4,454,7,6]})
b=pd.DataFrame({"a":[1,2,3,4,4],"b":[12,445,345,56,8],"c":[12,4,454,7,6]})
c=a.join(b)
print(c)
```
